Q9. Best allrounder performer (use bat avg >20 and bowling average in b/w 20 to 30




```python
df=pd.read_csv('/content/combined.csv')
AR_bat_avg=df[df['bat_avg']>=20]
AR_bw_avg=df[df['bw_avg']<30]
AR_bw_avg=AR_bw_avg[AR_bw_avg['bw_avg']>20]
```

```python
AR_df=pd.merge(AR_bat_avg,AR_bw_avg)
AR_df
```
